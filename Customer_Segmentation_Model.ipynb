```python
import numpy as np
import pandas as pd
df=pd.read_excel("superstore.xlsx")
df.head()
```

```python
df_flat=df.groupby("Customer_ID").agg(Customer_Name= ("Customer_Name","first"), Total_Spend=("Sales","sum")).sort_values(by="Total_Spend", ascending=False).reset_index()
df_flat.head(10)
```

```python
Total_Sales=df["Sales"].sum()
print(f"${Total_Sales:,.2f}")
```

```python
Pareto = (df_flat["Total_Spend"] / Total_Sales) * 100
print(Pareto.head(10))
```

```python
Total_Orders=df["Order_ID"].nunique()
print(Total_Orders)
```

```python
Latest_Purchase_Date=df[["Customer_Name","Order_Date"]].sort_values(by="Order_Date",ascending=False).head(1)
print(Latest_Purchase_Date)
```

```python
AOV=Total_Sales/Total_Orders
print(f"${AOV:.2f}")
```

```python
Running_Total=Pareto.cumsum()
print(Running_Total)
```

```python
import pandas as pd
pd.set_option("display.float_format", lambda x: "%.2f" % x)
df_flat["Pareto_Percentage"] = Running_Total
df_flat["Spend_Percentage_Of_Total"] = Pareto
df_flat["Total_Orders"] = Total_Orders
df_flat["Average_Order_Value"] = AOV
df_flat["Total_Sales"] = Total_Sales
df_flat
```

```python
df_flat["Orders_per_Customer"]=df.groupby("Customer_ID")["Order_ID"].nunique().values
df_flat.head()
```

```python
df_flat["Last_Purchase"]=(df.groupby("Customer_ID")["Order_Date"].max().values)
df_flat
```

```python
latestdate=df["Order_Date"].max()
print(latestdate)
```

```python
df_flat["Days_Without_Ordering"]=(latestdate-df_flat["Last_Purchase"]).dt.days
df_flat
```

```python
condition=[(df_flat['Orders_per_Customer'] >= 6) & (df_flat['Days_Without_Ordering'] <= 180),
             (df_flat['Orders_per_Customer'] >= 3) & (df_flat['Orders_per_Customer'] < 6) & (df_flat['Days_Without_Ordering'] <= 180),
           (df_flat['Orders_per_Customer'] >= 3) & (df_flat['Days_Without_Ordering'] > 180),
           (df_flat['Orders_per_Customer'] < 3) & (df_flat['Days_Without_Ordering'] <= 180)
          ]
choices=['Frequent Flyer (VIP)', 'Regular Customer', 'High-Value Churn Risk', 'New / Occasional']
df_flat["Customer Segment"]=np.select(condition, choices, default='Sleeper / Inactive')
df_flat["Customer Segment"].value_counts()
```

```python
df_flat
```

```python
import pandas as pd
df_flat["R_Score"] = pd.qcut(
    df_flat["Days_Without_Ordering"], q=5, labels=[5, 4, 3, 2, 1]
)
df_flat["F_Score"] = pd.qcut(
    df_flat["Orders_per_Customer"], q=5, labels=[1, 2, 3, 4, 5], duplicates="drop"
)
df_flat["M_Score"] = pd.qcut(
    df_flat["Total_Spend"], q=5, labels=[1, 2, 3, 4, 5]
)
```

```python
df_flat
```

```python
df_flat['RFM_Score'] = df_flat['R_Score'].astype(str) + df_flat['F_Score'].astype(str) + df_flat['M_Score'].astype(str)
```

```python
df_flat
```

```python
df_final = pd.merge(df, df_flat, on=["Customer_ID", "Customer_Name"], how="left")
df_final.head()
```

```python
df_flat.to_excel("Customer_Results_df_flat.xlsx", index=False)
df_final.to_excel("Customer_Results_df_final.xlsx", index=False)
```
